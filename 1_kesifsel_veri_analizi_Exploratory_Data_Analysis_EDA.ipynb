```python
# Veri analizi ve görselleştirme işlemleri için gerekli kütüphaneleri içe aktarıyoruz
import pandas as pd  # Veri manipülasyonu için Pandas
import matplotlib.pyplot as plt  # Grafik oluşturmak için Matplotlib
import seaborn as sns  # Daha estetik grafikler için Seaborn
```

```python
# Veri setini belirtilen dosya yolundan pandas ile okuyoruz
df = pd.read_csv(r"C:\Users\asli_\PYTHON_PROJECTS\pusula\side_effect_data.csv")
```

```python
# Veri Tiplerinin Doğrulanması ve Dönüşümü

df['Dogum_Tarihi'] = pd.to_datetime(df['Dogum_Tarihi'])
df['Ilac_Baslangic_Tarihi'] = pd.to_datetime(df['Ilac_Baslangic_Tarihi'])
df['Ilac_Bitis_Tarihi'] = pd.to_datetime(df['Ilac_Bitis_Tarihi'])
df['Yan_Etki_Bildirim_Tarihi'] = pd.to_datetime(df['Yan_Etki_Bildirim_Tarihi'])
```

```python
df.info()
```

```python
# Veri setindeki her sütundaki eksik değerlerin sayısını görmek için isnull().sum() fonksiyonunu kullanıyoruz
df.isnull().sum()
```

```python
# Kategorik değişkenlerde eksik değerleri en sık görülen değerle dolduruyoruz
df['Cinsiyet'] = df['Cinsiyet'].fillna(df['Cinsiyet'].mode()[0])
df['Il'] = df['Il'].fillna(df['Il'].mode()[0])
df['Alerjilerim'] = df['Alerjilerim'].fillna(df['Alerjilerim'].mode()[0])
df['Kronik Hastaliklarim'] = df['Kronik Hastaliklarim'].fillna(df['Kronik Hastaliklarim'].mode()[0])
df['Baba Kronik Hastaliklari'] = df['Baba Kronik Hastaliklari'].fillna(df['Baba Kronik Hastaliklari'].mode()[0])
df['Anne Kronik Hastaliklari'] = df['Anne Kronik Hastaliklari'].fillna(df['Anne Kronik Hastaliklari'].mode()[0])
df['Kiz Kardes Kronik Hastaliklari'] = df['Kiz Kardes Kronik Hastaliklari'].fillna(df['Kiz Kardes Kronik Hastaliklari'].mode()[0])
df['Erkek Kardes Kronik Hastaliklari'] = df['Erkek Kardes Kronik Hastaliklari'].fillna(df['Erkek Kardes Kronik Hastaliklari'].mode()[0])
df['Kan Grubu'] = df['Kan Grubu'].fillna(df['Kan Grubu'].mode()[0])

# Sayısal değişkenlerde eksik değerleri ortalama ile dolduruyoruz
df['Kilo'] = df['Kilo'].fillna(df['Kilo'].mean())
df['Boy'] = df['Boy'].fillna(df['Boy'].mean())
```

```python
df.isnull().sum()
#kontrol
```

```python
# Kategorik Değişkenlerin Analizi

sns.countplot(data=df, x='Cinsiyet')
plt.title('Cinsiyet Dağılımı')
plt.show()
```

```python
sns.countplot(data=df, x='Uyruk')
plt.title('Uyruk Dağılımı')
plt.show()
```

```python
sns.countplot(data=df, x='Kan Grubu')
plt.title('Kan Grubu Dağılımı')
plt.show()
```

```python
# Yan Etki ve İlaçlarla İlgili Daha Derinlemesine Analiz

yan_etki_by_ilac = df.groupby('Kan Grubu')['Yan_Etki'].value_counts().unstack().fillna(0)
sns.heatmap(yan_etki_by_ilac, cmap='Blues')
plt.title('İlaç ve Yan Etkiler Isı Haritası')
plt.show()
```

```python
# Yan Etki ve İlaçlarla İlgili Daha Derinlemesine Analiz

yan_etki_by_ilac = df.groupby('Alerjilerim')['Yan_Etki'].value_counts().unstack().fillna(0)
sns.heatmap(yan_etki_by_ilac, cmap='Blues')
plt.title('İlaç ve Yan Etkiler Isı Haritası')
plt.show()
```

```python
# Outlier (Aykırı Değer) Analizi

sns.boxplot(x=df['Kilo'])
plt.title('Kilo Aykırı Değer Analizi')
plt.show()
sns.boxplot(x=df['Boy'])
plt.title('Boy Aykırı Değer Analizi')
plt.show()
```

```python
#Kategorik Değişkenlerle Sayısal Değişkenlerin İlişkisi

sns.boxplot(x='Cinsiyet', y='Kilo', data=df)
plt.title('Cinsiyet ve Kilo İlişkisi')
plt.show()
sns.boxplot(x='Cinsiyet', y='Boy', data=df)
plt.title('Cinsiyet ve Boy İlişkisi')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Yan Etki Bildirim Tarihi sütununu datetime formatına çeviriyoruz
df['Yan_Etki_Bildirim_Tarihi'] = pd.to_datetime(df['Yan_Etki_Bildirim_Tarihi'])

# Yan etki bildirimlerinin zaman serisini oluşturuyoruz
yan_etki_zaman = df.groupby(df['Yan_Etki_Bildirim_Tarihi'].dt.date)['Yan_Etki'].count()

# Zaman serisini çiziyoruz
yan_etki_zaman.plot()

# Başlık ekliyoruz
plt.title('Yan Etki Bildirimleri Zaman Serisi')

# X eksenindeki tarihleri dikey yapmak için xticks'i 90 derece döndürüyoruz
plt.xticks(rotation=90)

# Grafiği gösteriyoruz
plt.show()
```

```python
# Matplotlib kütüphanesini içe aktarıyoruz
import matplotlib.pyplot as plt

# Veri setindeki sayısal değişkenler için histogram oluşturuyoruz
df.hist(bins=30, figsize=(10, 8))  # 30 aralıklı histogram, 10x8 inç boyutunda
plt.suptitle('Histogram Grafiği')  # Başlık ekliyoruz
plt.show()  # Grafiği gösteriyoruz
```

```python
# Seaborn kütüphanesini içe aktarıyoruz
import seaborn as sns

# Veri setindeki sayısal değişkenlerin dağılım grafiğini oluşturuyoruz
sns.pairplot(df)  # Tüm sayısal değişkenler arasındaki ilişkiyi gösteren grafik
plt.suptitle('Dağılım Grafikleri', y=1.02)  # Başlık ekliyoruz
plt.show()  # Grafiği gösteriyoruz
```

```python
# Kullanıcı ID'sini çıkararak yalnızca 'Kilo' ve 'Boy' sütunlarını seçiyoruz
df_filtered = df[['Kilo', 'Boy']]

# Kilo ve Boy arasındaki korelasyonu hesaplıyoruz
correlation_matrix = df_filtered.corr()

# Korelasyon ısı haritasını oluşturuyoruz
plt.figure(figsize=(6, 4))  # Grafik boyutunu ayarlıyoruz
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')  # Isı haritasını çiziyoruz
plt.title('Kilo ve Boy Korelasyon Isı Haritası')  # Başlık ekliyoruz
plt.show()  # Grafiği gösteriyoruz
```
